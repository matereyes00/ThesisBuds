```python
import pandas as pd
import geotable
import geopandas as gpd
from geopandas import GeoDataFrame
```

```python
pois_df = gpd.read_file("sablayan-pois-with-brgy.geojson")
pois_df.head()
```

```python
roads_df = gpd.read_file("sablayan-roads-with-brgy.geojson")
roads_df.head()
```

## Cleaning Road Network Data

```python
roads_df
```

```python
# get the column of brgys
bc = roads_df["brgys"]
```

```python
# Separating the psgc and brgy names from the dict values
psgc = []
brgy = []
for i in range(1013):
  # key, value = list(bc.items())[i]
  psgc.append(list(bc[i].keys())[0])
  brgy.append(list(bc[i].values())[0])


print(psgc)
print(brgy)
```

```python
roads_df['psgc'] = ''
```

```python
# replacing values in columns psgc and brgys with the list values
for i in range(1013):
  roads_df['psgc'][i] = psgc[i]
  roads_df['brgys'][i] = brgy[i]

print(roads_df['psgc'])
print(roads_df['brgys'])
```
